```python
import pandas as pd
df = pd.read_csv("drive/MyDrive/BRIN_Dataset/inacoved.csv")
df
```

```python
df = df.drop('title', axis=1)
df = df.rename(columns={'title clean' : 'title'})

df['published_at_iso'] = pd.to_datetime(df['published_at_iso'])
df['date_published'] = df['published_at_iso'].dt.date
df['time_published'] = df['published_at_iso'].dt.time

df = df.drop('published_at_iso', axis=1)
```

```python
# Mengubah kolom 'date_published' menjadi tipe data datetime
df['date_published'] = pd.to_datetime(df['date_published'])
```

```python
df_filtered = df[df['date_published'] >= '2020-01-01']
df_filtered = df_filtered[df_filtered['date_published'] < '2020-05-23']

df_filtered
```

```python
df_filtered['event_daily_news'] = df_filtered.groupby('date_published')['event'].transform('sum')

df_result = df_filtered[['date_published', 'event_daily_news']].drop_duplicates()
df_result
```

```python
# # Menetapkan 'date_published' sebagai indeks DataFrame
df_filtered.set_index(['date_published'], inplace=True)
```

```python
import matplotlib.pyplot as plt

# Meresample data ke tingkat harian ('D') dan menjumlahkan nilai 'event' per hari
df_filtered = df_filtered['event'].resample('D').sum()

# Mengonversi hasil resample ke dalam DataFrame baru
df_filtered = pd.DataFrame(df_filtered)

# Mengatur ukuran gambar plot
plt.figure(figsize=(15, 5))

# Membuat plot garis untuk kolom 'event'
df_filtered['event'].plot()

# Menambah label sumbu x (Date) dan sumbu y (Number of Event Published)
plt.xlabel('Month')
plt.ylabel('Sum of Event')

# Menampilkan plot
plt.show()
```
